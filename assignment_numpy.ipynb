**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
import numpy as np
list=np.random.randint(1,10,20)
result=np.where((list>=3) & (list<=8),list*(-1),list)
print(result)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
import numpy as np
list=np.random.randint(1,10,20)
print(list)
result=np.where(list==list.max(),list*0,list)
print(result)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
arr = [[1, 2], [3, 4, 5]]
meshes = [np.array(subarr) for subarr in arr]
result = np.array(np.meshgrid(*meshes)).T.reshape(-1, len(arr))
for combination in result:
    print(combination)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(1, 6, size=(8, 3))
B = np.random.randint(1, 6, size=(2, 2))
print(A)
print(B)
result_indices = []
for row in B:
    row_indices = np.any(np.isin(A, row).reshape(-1, 3), axis=1)
    result_indices.append(row_indices)

rows_indices = np.all(result_indices, axis=0)

if np.any(rows_indices):
    result = A[rows_indices]
else:
    result = None
print("Строки в A, содержащие элементы из каждой строки в B:")
print(result)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
random_matrix = np.random.randint(1, 4, size=(10, 3))
print(random_matrix)
rows_with_unique_values = random_matrix[np.any(np.diff(random_matrix, axis=1), axis=1)]
print('\n')
print(rows_with_unique_values)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# numpy
x = np.random.randint(0, 3, 8)
print(x)
changes = np.flatnonzero(np.diff(x) != 0)
changes = np.append(changes, len(x) - 1)
values = x[changes]
lengths = np.diff(np.concatenate(([0], changes + 1)))
print("Значения:", values)
print("Длины серий:", lengths)
#без numpy
p = []
numbers = []
count_of_numbers = []
prev = None
counter = 0
for i in range(0, 8):
    p.append(randint(1, 5))
print(p)
for i in range(0, 8):
    if p[i] != prev:
        if prev is not None:
            numbers.append(prev)
            count_of_numbers.append(counter)
        prev = p[i]
        counter = 0
    counter += 1
if prev is not None:
    numbers.append(prev)
    count_of_numbers.append(counter)
print("Значения:",numbers)
print("Длины серий:",count_of_numbers)
```
